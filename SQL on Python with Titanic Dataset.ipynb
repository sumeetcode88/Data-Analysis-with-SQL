### WHERE Clause


WHERE clause used to pull rows based on condition given.
Here we are asking SQL to pull rows where Passenger age is less than 15

```python
ps.sqldf("""

SELECT df.*
FROM titanic_disaster_dataset AS df
WHERE df.age < 15

""")
```

Sometimes the data we recieve has missing values and we need to treat these values for further analysis. In SQL we use IS NULL keyword to find Null values and NOT NULL for Non-Null values.

```python
ps.sqldf("""

SELECT df.*
FROM titanic_disaster_dataset AS df
WHERE df.age IS NULL

""")
```

```python
ps.sqldf("""

SELECT df.*
FROM titanic_disaster_dataset AS df
WHERE df.age NOT NULL

""")
```

Use BETWEEN to filter numerical values of rows. Here we have pull all passengers between the age of 5 and 10

```python
ps.sqldf("""

SELECT 
    df.*
FROM titanic_disaster_dataset AS df
WHERE df.age BETWEEN 5 AND 10

""")
```

With WHERE clause if multiple conditions are to be passed we use "AND" or "OR" or combination of these two.

"CASE WHEN" is used for creating a new column based on conditions we pass. In our data have AGE column having differnt age groups, let us check how we can create a column for it. We have used IN to make sure if age group are defined correctly.

```python
ps.sqldf("""

SELECT 
    df.Name, df.PassengerID, df.Age,
    CASE WHEN df.Age IS NULL THEN 'Missing'
        WHEN df.Age <12 THEN 'Kid'
        WHEN df.Age <20 THEN 'Teenage'
        WHEN df.Age <60 THEN 'Adult'
    ELSE 'Senior Citizen' END AS Age_Group
FROM titanic_disaster_dataset AS df
WHERE df.Age IN (5,15,25,65)
LIMIT 10   
    
""")
```

### AGGREGATIONS

Lets look at the aggregation of data. We have tried commonly used functions like SUM, AVG, MIN, MAX, COUNT on Age column for understanding.

```python
ps.sqldf("""

SELECT 
    AVG(df.Age) AS avg_age,
    SUM(df.Age) AS tot_age,
    MIN(df.Age) AS min_age,
    MAX(df.Age) AS max_age,
    COUNT(df.PassengerId) AS tot_passengers,
    COUNT(DISTINCT df.Ticket) AS dist_tickets
FROM titanic_disaster_dataset AS df

""")
```

If we look at above results it did calculations, but on all rows at once. This result may not be usable. What if we do aggregation particularly based on anyone column like Embarked which has 4 unique values. So this is example of GROUP BY clause which is used to combine rows on the basis of column values i.e. we will get 4 rows and each row will be aggregated value for Emarked value.

```python
ps.sqldf("""

SELECT 
    DISTINCT df.Embarked
FROM titanic_disaster_dataset AS df
    
""")
```

```python
ps.sqldf("""

SELECT 
    df.Embarked,
    AVG(df.Age) AS Avg_Age,
    SUM(df.Age) AS Sum_Age,
    MIN(df.Age) AS Min_Age,
    MAX(df.Age) AS Max_Age,
    COUNT(df.PassengerId) AS Total_passengers,
    COUNT(DISTINCT df.Ticket) AS Unique_tickets
FROM titanic_disaster_dataset AS df
GROUP BY df.Embarked

""")
```

In GROUP BY clause we can pass multiple columns, which will create output of unique combination of values of these multiple columns. See this example where we have also combined WHERE.

```python
ps.sqldf("""

SELECT 
    df.Embarked, df.Sex,
    AVG(df.Age) AS Avg_Age,
    SUM(df.Age) AS Sum_Age,
    MIN(df.Age) AS Min_Age,
    MAX(df.Age) AS Max_Age,
    COUNT(df.PassengerId) AS Total_passengers,
    COUNT(DISTINCT df.Ticket) AS Unique_tickets
FROM titanic_disaster_dataset AS df

WHERE df.Age>18 and df.Age<60

GROUP BY df.Embarked, df.Sex

""")
```

We can also sort the data using ORDER BY. For an example lets use same query as above and add ORDER BY clause. By default it gives result in Ascending order and for Descending order we use DESC at the end.

```python
ps.sqldf("""

SELECT 
    df.Embarked, df.Sex,
    AVG(df.Age) AS Avg_Age,
    SUM(df.Age) AS Sum_Age,
    MIN(df.Age) AS Min_Age,
    MAX(df.Age) AS Max_Age,
    COUNT(df.PassengerId) AS Total_passengers,
    COUNT(DISTINCT df.Ticket) AS Unique_tickets
FROM titanic_disaster_dataset AS df

WHERE df.Age>18 and df.Age<60

GROUP BY df.Embarked, df.Sex

ORDER BY Min_Age Desc
""")
```

On the aggregated information we can put filter by using HAVING clause. The difference between WHERE and HAVING is that WHERE is used on data directly from dataset and HAVING is used on aggregated data.

```python
ps.sqldf("""

SELECT 
    df.Embarked, df.Sex,
    AVG(df.Age) AS Avg_Age,
    SUM(df.Age) AS Sum_Age,
    MIN(df.Age) AS Min_Age,
    MAX(df.Age) AS Max_Age,
    COUNT(df.PassengerId) AS Total_passengers,
    COUNT(DISTINCT df.Ticket) AS Unique_tickets
FROM titanic_disaster_dataset AS df

WHERE df.Age>18 and df.Age<60

GROUP BY df.Embarked, df.Sex

HAVING Avg_Age > 30

ORDER BY df.Embarked, df.Sex
""")
```

<b>INNER JOIN</b><br>
INNER JOIN gives rows corresponding to only the common values(in the columns used in the inner join condition) between the two tables

JOIN keyword by default is INNER JOIN. Below query shows as common values between two tables "titanic_disaster_dataset" and "cities". 
<br>"S" and "C" are picked and "Q" is missed.

```python
ps.sqldf("""

SELECT 
    c.city,
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS Avg_Age,
    SUM(df.Age) AS Sum_Age,
    MIN(df.Age) AS Min_Age,
    MAX(df.Age) AS Max_Age,
    COUNT(df.PassengerId) AS Total_passengers,
    COUNT(DISTINCT df.Ticket) AS Unique_tickets
    
FROM titanic_disaster_dataset AS df

INNER JOIN cities AS c
    ON df.Embarked = c.code

GROUP BY df.Embarked,
    df.Sex

""")
```

<b>LEFT JOIN</b><br>
LEFT JOIN gives all rows from the left table and only those rows from right table where the values are common in the columns used in left join condition

The table "titanic_disaster_dataset" is on LEFT and below query pulls all values of this table whereas the values not found in "cities" are showing as NULL(None).

```python
ps.sqldf("""

SELECT 
    c.city,
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS Avg_Age,
    SUM(df.Age) AS Sum_Age,
    MIN(df.Age) AS Min_Age,
    MAX(df.Age) AS Max_Age,
    COUNT(df.PassengerId) AS Total_passengers,
    COUNT(DISTINCT df.Ticket) AS Unique_tickets
    
FROM titanic_disaster_dataset AS df

LEFT JOIN cities AS c
    ON df.Embarked = c.code

GROUP BY df.Embarked,
    df.Sex

""")
```

<b>RIGHT JOIN</b><br>
RIGHT JOIN is the mirror opposite of LEFT JOIN. RIGHT JOIN gives all rows from the right table and only those rows from left table where the values are common in the columns used in right join condition.

pandasql do not support RIGHT JOIN but if we swap the tables we will get same output as RIGHT JOIN. Check below query, all values of "cities" table are pulled in the results.

```python
ps.sqldf("""

SELECT 
    c.city,
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS Avg_Age,
    SUM(df.Age) AS Sum_Age,
    MIN(df.Age) AS Min_Age,
    MAX(df.Age) AS Max_Age,
    COUNT(df.PassengerId) AS Total_passengers,
    COUNT(DISTINCT df.Ticket) AS Unique_tickets
    
FROM cities AS c

LEFT JOIN titanic_disaster_dataset AS df
ON df.Embarked = c.code

GROUP BY df.Embarked,
    df.Sex

""")
```

<b>FULL JOIN</b><br>
FULL JOIN gives all rows from the from both tables, irrespective of common values

pandasql do not support FULL JOIN too. The output has both common and non-common elements from both the tables.

```python
ps.sqldf("""

SELECT 
    c.city,
    df.Embarked,
    df.Sex,
    AVG(df.Age) AS Avg_Age,
    SUM(df.Age) AS Sum_Age,
    MIN(df.Age) AS Min_Age,
    MAX(df.Age) AS Max_Age,
    COUNT(df.PassengerId) AS Total_passengers,
    COUNT(DISTINCT df.Ticket) AS Unique_tickets
    
FROM cities AS c

FULL OUTER JOIN titanic_disaster_dataset AS df
ON df.Embarked = c.code

GROUP BY df.Embarked,
    df.Sex

""")
```

## Ranking in SQL

#### RANK(), DENSE_RANK() and ROW_NUMBERS() are three functions which help you to sort the data and give rank to each row.<br>
Generally these functions are used with OVER() function to sort the data based on some column before giving rank to each row and also to partition the data.

RANK() provides a rank to each row. If any n rows have the same value in the sorting column, then RANK() gives the same value to all of these rows. For the next row where the value is different than the previous n rows, the rank given will be previous rank plus n.<br>
Notice row no. 2 and 3, where the age values are the same(i.e. 71) and hence both rows have the same rank(i.e. 3). In row 4, the ranking begins from 5(i.e. previous rank 3 plus 2) and not 4.

```python
ps.sqldf("""

SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    RANK() OVER(ORDER BY df.Age DESC) AS Rank
FROM titanic_disaster_dataset AS df

""")
```

RANK() will skip rank if same values found in the sorting order.<br>
To tackle this we can use DENSE_RANK() to ensure that no ranks are skipped.<br>
Take a look at the example below, where row no 4 has a rank 4 and not 5.

```python
ps.sqldf("""

SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    DENSE_RANK() OVER(ORDER BY df.Age DESC) AS Rank
FROM titanic_disaster_dataset AS df

""")
```

Now, if suppose we do not want to repeat ranks and it should continue ranking on values. Here we can use ROW_NUMBER() which gives different rank to each row even if the sorted values are same.<br>
Notice below row no 2 and 3 which has differnt ranks even though Age is same.

```python
ps.sqldf("""

SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    ROW_NUMBER() OVER(ORDER BY df.Age DESC) AS Rank
FROM titanic_disaster_dataset AS df

""")
```

We have used ORDER BY clause on above query but with single column. Can we use multiple columns and specify sorting order, the answer is YES.

```python
ps.sqldf("""

SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    RANK() OVER(ORDER BY df.Age DESC, df.Fare ASC) AS rnk
    
FROM titanic_disaster_dataset AS df

""")
```

Previously i had mentioned that OVER() is used to sort the data based on some column before giving rank to each row and it can be used to partition the data too.<br>
Suppose you want to give separate ranks withtin specific levels of data, then you can use "PARTITION BY" in OVER() function. In the example below, for each combination of df.Embarked and df.Sex, we will get different sets of ranks. Refer row no 2 and 3, a different ranking is applied.

```python
ps.sqldf("""

SELECT 
    df.Name,
    df.Embarked,
    df.Sex,
    df.Age,
    df.Fare,
    RANK() OVER(PARTITION BY df.Embarked, df.Sex 
                ORDER BY df.Age DESC, df.Fare DESC) AS rnk
    
FROM titanic_disaster_dataset AS df

""")
```

<b>Lead and Lag</b><br>
LEAD() and LAG() can be used to get the nth previous or nth next value in a column, after sorting the data using OVER()<br>
In these functions the third parameter is returned in result when previous/next value is not found,. If you do not mention third parameter it will return NULL.

```python
ps.sqldf("""

SELECT 
    df.Name,
    LEAD(df.Name, 1,0) OVER(ORDER BY df.Name) AS next_name,
    LEAD(df.Name, 2, 0) OVER(ORDER BY df.Name) AS second_next_name,
    LAG(df.Name, 1, 0) OVER(ORDER BY df.Name) AS previous_name,
    LAG(df.Name, 2, 0) OVER(ORDER BY df.Name) AS second_previous_name
    
FROM titanic_disaster_dataset AS df

""")
```

Using "PARTITION BY" within OVER() function, you can partition the data and get LEAD() and LAG() of a column within partitions

```python
ps.sqldf("""
SELECT df.Name, df.Embarked,
LEAD(df.Name, 1) OVER(PARTITION BY df.Embarked ORDER BY df.Name DESC) AS df_name
FROM titanic_disaster_dataset AS df
---ORDER BY df.Embarked
LIMIT 5
""")
```

```python
ps.sqldf("""

SELECT 
    df.Name,
    df.Embarked,
    LEAD(df.Name, 1,0) OVER(PARTITION BY df.Embarked ORDER BY df.Name DESC) AS next_name,
    LEAD(df.Name, 2, 0) OVER(PARTITION BY df.Embarked ORDER BY df.Name) AS second_next_name,
    LAG(df.Name, 1, 0) OVER(PARTITION BY df.Embarked ORDER BY df.Name) AS previous_name,
    LAG(df.Name, 2, 0) OVER(PARTITION BY df.Embarked ORDER BY df.Name) AS second_previous_name
    
FROM titanic_disaster_dataset AS df
LIMIT 15
""")
```

<b>Subqueries</b><br>
Subqueries are small queries which are used within a bigger main SQL query. Subqueries store the output of a SQL temporarily and more SQL queries can be written on top of these.

```python
ps.sqldf("""

SELECT df1.Embarked, df1.avg_age
    
FROM (SELECT a.Embarked, AVG(a.Age) AS avg_age 
      FROM titanic_disaster_dataset AS a
      GROUP BY a.Embarked) AS df1

WHERE df1.avg_age<30
    
""")
```

Let's break the above query. Below output is for subquery. We are pulling data fro this output on the condition that avg_age should be less than 30. This is what we have received in above output.

```python
ps.sqldf("""
SELECT a.Embarked, AVG(a.Age) AS avg_age 
      FROM titanic_disaster_dataset AS a
      GROUP BY a.Embarked
""")
```

Let's look at one more example...<br>
We will pull average age on Embarked column wise<br>
then we will pull total passengers on Embarked column wise<br>
Further, we will combine these 2 results with JOIN to get common values

```python
ps.sqldf("""
SELECT a.Embarked, AVG(a.Age) AS avg_age 
      FROM titanic_disaster_dataset AS a
      GROUP BY a.Embarked    
""")
```

```python
ps.sqldf("""

SELECT b.Embarked, COUNT(b.PassengerID) AS tot_passengers
     FROM titanic_disaster_dataset AS b
     GROUP BY b.Embarked

""")
```

```python
ps.sqldf("""

SELECT df1.Embarked, df1.avg_age, df2.tot_passengers
FROM (SELECT a.Embarked, AVG(a.Age) AS avg_age 
      FROM titanic_disaster_dataset AS a
      GROUP BY a.Embarked) AS df1

INNER JOIN

    (SELECT b.Embarked, COUNT(b.PassengerID) AS tot_passengers
     FROM titanic_disaster_dataset AS b
     GROUP BY b.Embarked) AS df2

ON df1.Embarked = df2.Embarked
    
""")
```

<b> Nested queries </b><br>
We can use nested subqueries inside subquery and also use it to filter the data.<br>
We are running subqueries separetly to check how the query come up with the final output.

```python
ps.sqldf("""
SELECT b.* FROM titanic_disaster_dataset AS b
WHERE b.Embarked IN 
       (SELECT DISTINCT df1.Embarked
        FROM (SELECT a.Embarked, AVG(a.Age) AS avg_age 
              FROM titanic_disaster_dataset AS a
              GROUP BY a.Embarked) AS df1
        WHERE df1.avg_age>30)
LIMIT 5
""")
```

```python
ps.sqldf("""
SELECT DISTINCT df1.Embarked
        FROM (SELECT a.Embarked, AVG(a.Age) AS avg_age 
              FROM titanic_disaster_dataset AS a
              GROUP BY a.Embarked) AS df1
        WHERE df1.avg_age>30
        """)
```

So below we have all values of Embarked column with average age. We pull avg_age on condition that it should be > 30 which gave only "C" as output. So our final output has rows which has Emabrked with values "C".
